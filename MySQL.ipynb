### LIKE OPERATOR

### Aliase 

```python
cursor.execute('USE world')

query = """
SELECT SUM(Population) AS Pop_sum
FROM city
"""

data = pd.read_sql_query(query, myconn)

data
```

```python
query = """
SELECT 
SUM(Population) AS Pop_sum,
MIN(Population) AS Pop_min,
MAX(Population) AS Pop_max,
COUNT(Population) AS Pop_count


FROM city
"""

data = pd.read_sql_query(query, myconn)

data
```

```python
query = """
SELECT 
SUM(Population) AS Pop_sum,
MIN(Population) AS Pop_min,
MAX(Population) AS Pop_max,
COUNT(Population) AS Pop_count

FROM city

WHERE Population > 100000
"""

data = pd.read_sql_query(query, myconn)

data
```

```python
query = """
SELECT 
SUM(Population) AS Pop_sum,
MIN(Population) AS Pop_min,
MAX(Population) AS Pop_max,
COUNT(Population) AS Pop_count

FROM city

WHERE Population BETWEEN 100000 AND 2000000
"""

data = pd.read_sql_query(query, myconn)

data
```
