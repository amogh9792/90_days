```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

pd.pandas.set_option("display.max_columns", None)
#pd.pandas.set_option("display.max_rows", None)
```

```python
data = pd.read_csv("house_price.csv")

data.head()
```

```python
data.tail()
```

```python
data.shape
```

### Missing values

```python
data.isnull().sum()
```

```python
feature_with_na = [feature for feature in data.columns if data[feature].isnull().sum() > 0]

feature_with_na
```

#### Total count of feature with null values

```python
print(len(feature_with_na))
```

#### Find the percentage of missing values in columns

```python
for feature in feature_with_na:
    
    col_na_percentage = data[feature].isnull().mean()*100
    print(f"Feature : {feature} NA Value Percentage : {col_na_percentage}")
```

```python
data.shape
```

### BoxPlot

```python
sns.boxplot(data.SalePrice)
```

```python
sns.boxplot(data.BsmtUnfSF)
```

```python
df = data.copy()

for feature in feature_with_na:
    
    df[feature] = np.where(df[feature].isnull(), 1, 0)

    df.groupby(feature)['SalePrice'].median().plot.bar()
    
    plt.title(feature)
    
    plt.show()
```

### Temporal variables 

```python
year_feature = [feature for feature in data.columns if 'Yr' in feature or 'Year' in feature]

year_feature
```

```python
for feature in year_feature:
    
    print(feature, data[feature].unique())
```

```python
df = data.copy()

df.groupby('YrSold')['SalePrice'].median().plot()

plt.xlabel("Year sold")
plt.ylabel("Sale Price")
plt.title("House Price VS Year Sold")
plt.show()
```

```python
df = data.copy()

df.groupby('YearBuilt')['SalePrice'].median().plot()

plt.xlabel("Year YearBuilt")
plt.ylabel("Sale Price")
plt.title("House Price VS Year YearBuilt")
plt.show()
```

```python
df = data.copy()

df.groupby('YearBuilt')['SalePrice'].median().plot()

plt.xlabel("Year YearBuilt")
plt.ylabel("Sale Price")
plt.title("House Price VS Year YearBuilt")
plt.show()
```

```python
df = data.copy()

for feature in year_feature:
    
    if feature != 'YrSold':
        
        df[feature] = df['YrSold'] - df[feature]
        
        plt.scatter(df[feature], df['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Sale Price')
        plt.show()
```
