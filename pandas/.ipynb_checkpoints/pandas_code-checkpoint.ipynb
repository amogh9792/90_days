### Grouping

```python
data.groupby("Cabin").sum()['Fare']
```

```python
data.groupby('Sex').sum()['Fare']
```

```python
def apply_grp(x):
    
    if x <= 100:
        return "A"
    elif x > 100 and x <= 200:
        return "B"
    else:
        return "C"
    
data['Fare_Group'] = data['Fare'].apply(apply_grp)
```

```python
data
```

```python
data.groupby('Fare_Group').max()['Fare']
```

### Merge Dataframes

```python
a1 = {
    
    "a": [1,2,3,4],
    "b": [88,44,55,76],
    "c": [12,34,65,77]
}
```

```python
data1 = pd.DataFrame(a1)
```

```python
data1
```

```python
a2 = {
    
    "a": [1,2,3,4],
    "b": [12,"dkjf",22,22],
    "c": [33,54,"jjg",22]
}
```

```python
data2 = pd.DataFrame(a2)
```

```python
data2
```

```python
data1
```

```python
data2
```

```python
pd.merge(data1,data2, how = "outer", on = ['a'])
```
